## 人口变化及占比

```python
import pandas as pd

# 读取CSV文件
file_path = "./data/乡村和城镇人口数据.csv"
data = pd.read_csv(file_path)

# 显示数据的前几行
print(data.head())
```

## 堆积面积图

```python
# 设置年份作为索引
data.set_index('指标', inplace=True)

# 提取所需数据
total_population = data.loc['年末总人口(万人)']
rural_population = data.loc['乡村人口(万人)']

# 倒置数据，以便堆叠
total_population = total_population[::-1]
rural_population = rural_population[::-1]


# 定义颜色
colors = ['orange', 'skyblue']

# 绘制堆积面积图，并设置颜色
plt.figure(figsize=(10, 6))
plt.stackplot(total_population.index, rural_population, total_population - rural_population, colors=colors, labels=['乡村人口', '城镇人口'])
plt.xlabel('年份')
plt.ylabel('人口数量（万人）')
plt.title('年末总人口与乡村人口变化趋势')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 乡村和城镇人口数据

```python
import pandas as pd

# 读取CSV文件
file_path = "./data/城乡就业人员数据.csv"
data = pd.read_csv(file_path)

# 显示数据的前几行
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取CSV文件
file_path = "./data/城乡就业人员数据.csv"
data = pd.read_csv(file_path)

# 提取所需数据
years = data.columns[1:].astype(str)
employment = data.loc[0, years]
urban_employment = data.loc[1, years]
rural_employment = data.loc[2, years]
private_urban_employment = data.loc[3, years].fillna(0)
individual_urban_employment = data.loc[4, years].fillna(0)

# 计算总就业人员数
total_employment = employment 

# 设置每个条形的宽度
bar_width = 0.35

# 设置每个条形的位置
r1 = np.arange(len(years))

# 绘制堆积条形图
plt.figure(figsize=(10, 6))
plt.bar(r1, urban_employment, color='#FF8080', width=bar_width, edgecolor='grey', label='城镇就业人员')
plt.bar(r1, rural_employment, bottom=urban_employment, color='#FFCF96', width=bar_width, edgecolor='grey', label='乡村就业人员')
plt.bar(r1, private_urban_employment, bottom=urban_employment + rural_employment, color='y', width=bar_width, edgecolor='grey', label='私营企业乡村就业人员')
plt.bar(r1, individual_urban_employment, bottom=urban_employment + rural_employment + private_urban_employment, color='orange', width=bar_width, edgecolor='grey', label='个体乡村就业人员')
plt.plot(r1, total_employment, color='b', marker='o', linestyle='-', linewidth=2, label='总就业人员')

# 添加标签
plt.xlabel('年份', fontweight='bold')
plt.ylabel('就业人员(万人)', fontweight='bold')
plt.title('城乡就业人员数据', fontweight='bold')
plt.xticks(r1, years, rotation=45)
plt.legend()
plt.tight_layout()

# 反转 x 轴
plt.gca().invert_xaxis()

# 显示图形
plt.show()
```

1.总人数差不多

2.cz增多，xc下降

3.有些缺失数据

## 折线图

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
file_path = "./data/乡村基层组织数据.csv"
data = pd.read_csv(file_path)

# 提取数据
years = data.columns[1:].astype(str)
township_count = data.loc[0, years]
village_count = data.loc[1, years]
committee_count = data.loc[3, years]
town_count = data.loc[2, years]  # 添加镇数(个)

# 创建画布和两个子图
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# 绘制第一个子图（村民委员会数）
axs[0].plot(years[::-1], committee_count[::-1], marker='o', color='#FF9B9B', label='村民委员会数(个)')
axs[0].set_title('村民委员会数(个)变化趋势')
axs[0].set_xlabel('年份')
axs[0].set_ylabel('村民委员会数(个)')
axs[0].legend()

# 绘制第二个子图（其余三个指标）
axs[1].plot(years[::-1], township_count[::-1], marker='o', color='#9EA1D4', label='乡镇数(个)')
axs[1].plot(years[::-1], town_count[::-1], marker='o', color='#A8D1D1', label='镇数(个)')  # 添加镇数(个)的折线图
axs[1].plot(years[::-1], village_count[::-1], marker='o', color='#FD8A8A', label='乡数(个)')
axs[1].set_title('乡镇数(个)和乡数(个)变化趋势')
axs[1].set_xlabel('年份')
axs[1].set_ylabel('数量(个)')
axs[1].legend()

# 调整x轴刻度标签，避免重叠
plt.setp(axs[0].xaxis.get_majorticklabels(), rotation=45)
plt.setp(axs[1].xaxis.get_majorticklabels(), rotation=45)

# 自动调整子图布局
plt.tight_layout()

# 显示图形
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
file_path = "./data/主要农作物播种面积和产品产量数据.csv"
data = pd.read_csv(file_path)

# 提取数据
years = data.columns[1:].astype(str)
total_crop_area = data.loc[0, years]

# 创建画布和子图
fig, ax = plt.subplots(figsize=(10, 6))

# 绘制播种面积折线图
ax.plot(years, total_crop_area, marker='o', color='#FF8787', label='农作物总播种面积(千公顷)')

# 设置对数坐标轴
ax.set_yscale('log')

# 添加标题和标签
ax.set_title('农作物总播种面积变化趋势')
ax.set_xlabel('年份')
ax.set_ylabel('农作物总播种面积(千公顷)')
ax.legend()

# 调整x轴刻度标签，避免重叠
plt.xticks(rotation=45)

# 反转 x 轴
plt.gca().invert_xaxis()

# 显示图形
plt.tight_layout()
plt.show()
```

## 农村居民主要食品消费量数据


```python
import pandas as pd

# 读取 CSV 文件
file_path = "./data/农村居民主要食品消费量数据.csv"
data = pd.read_csv(file_path, encoding="ANSI")

# 显示数据的前几行
print(data.head())
```

```python
import matplotlib.pyplot as plt
import squarify

# 指标列表
indicators = data['指标']

# 年份
year = '2022年'

# 数据
consumption_data = data[year]

# 删除每个指标中的“村居民人均”、“消费量(千克)”
indicators = indicators.str.replace('村居民人均', '').str.replace('消费量\(千克\)', '', regex=True)

# 手动指定颜色列表
colors = ['#E72929', '#FF5BAE', '#FFE4CF', '#FFFDD7', '#00224D', '#FFAF45', '#B0C5A4', '#711DB0', '#468B97', '#5D9C59']

# 绘图
plt.figure(figsize=(12, 8))
squarify.plot(sizes=consumption_data, label=[f"{ind}: {con:.1f}" for ind, con in zip(indicators, consumption_data)], color=colors)
plt.axis('off')  # 关闭坐标轴
plt.title('2022年中国农村居民人均主要食品消费量(千克) Treemap')
plt.show()
```

## 农村居民平均每百户年末主要耐用消费品拥有量数据

```python
import pandas as pd

# 读取 CSV 文件
file_path = "./data/农村居民平均每百户年末主要耐用消费品拥有量数据.csv"
data = pd.read_csv(file_path, encoding="ANSI")

# 显示数据的前几行
print(data.head())
```

```python
import matplotlib.pyplot as plt

# 选择指标列
indicators = data['指标']

# 删除指标中的部分文字
indicators = indicators.str.replace('农村居民平均每百户年末', '')
indicators = indicators.str.replace('拥有量', '')

# 选择2022年的数据
data_2022 = data['2022年']

# 筛选出每百户人均数量低于一半（50）的指标
low_values = data_2022[data_2022 < 50]
high_values = data_2022[data_2022 >= 50]

# 对低值和高值进行排序
low_values_sorted = low_values.sort_values()
high_values_sorted = high_values.sort_values()

# 创建画布并设置子图布局
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# 左边图形，低于一半的指标排名
axs[0].barh(range(len(low_values_sorted)), low_values_sorted, color='skyblue')
axs[0].set_xlabel('每百户人均数量')
axs[0].set_yticks(range(len(low_values_sorted)))
axs[0].set_yticklabels(indicators[low_values_sorted.index])  # 设置y轴刻度为指标名称
axs[0].set_title('人均数量低于0.5的指标排名')

# 右边图形，高于一半的指标排名
axs[1].barh(range(len(high_values_sorted)), high_values_sorted, color='lightgreen')
axs[1].set_xlabel('每百户人均数量')
axs[1].set_yticks(range(len(high_values_sorted)))
axs[1].set_yticklabels(indicators[high_values_sorted.index])  # 设置y轴刻度为指标名称
axs[1].set_title('人均数量高于0.5的指标排名')

# 自动调整子图布局
plt.tight_layout()

# 显示图形
plt.show()
```

## 农村居民人均收支数据

```python
import pandas as pd

# 读取 CSV 文件
file_path = "./data/农村居民人均收支数据.csv"
data = pd.read_csv(file_path, encoding="ANSI")

# 显示数据的前几行
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 获取数据的最后5行
data_last_five = data.tail()

# 删除指标中的“家庭农村居民人均可支配收入(元)”部分
data_last_five['指标'] = data_last_five['指标'].str.replace('家庭农村居民人均可支配收入\(元\)', '').str.strip()

# 获取2022年和2012年的数据
data_2022 = data_last_five[['指标', '2022年']].dropna()
data_2012 = data_last_five[['指标', '2013年']].dropna()

# 创建画布和子图
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# 绘制2022年数据饼图
axes[0].pie(data_2022['2022年'], labels=data_2022['指标'], autopct=lambda p: f'{p:.1f}%\n({p * sum(data_2022["2022年"]) / 100:.0f}元)', startangle=140)
axes[0].set_title('2022年家庭农村居民人均可支配收入占比')
axes[0].axis('equal')  # 保持圆形

# 绘制2012年数据饼图
axes[1].pie(data_2012['2013年'], labels=data_2012['指标'], autopct=lambda p: f'{p:.1f}%\n({p * sum(data_2012["2013年"]) / 100:.0f}元)', startangle=140)
axes[1].set_title('2013年家庭农村居民人均可支配收入占比')
axes[1].axis('equal')  # 保持圆形

plt.show()
```

全国居民五等份收入分组是指将所有调查户按人均收入水平从低到高顺序排列，平均分为五个等份，处于最低20%的收入家庭为低收入组，依此类推依次为中间偏下收入组、中间收入组、中间偏上收入组、高收入组。

## 结合之前可以看出，在中国变化激烈的10年内，在五等份收入分组情况下，虽然收入变多了，但贫富差距没有变化（即没有变大、也没有变小，挺微妙的）
